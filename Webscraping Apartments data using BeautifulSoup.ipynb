I recently visited <a href = 'https://www.property24.co.ke/apartments-flats-to-rent'> Property24</a>; a house listing site and was amazed at the huge amount of data regarding different housing options. I then decided to scrape the flats and apartments data and this is what I came up with.

```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
from time import sleep
from random import randint
```

```python
title = []
location = []
price = []
rate = []
bedrooms = []
bathrooms =[]

pages = np.arange(1, 2507, 21)

for page in pages:
    url = 'https://www.property24.co.ke/apartments-flats-to-rent'
    page = requests.get(url)
    unsdata = BeautifulSoup(page.content, 'html.parser')
    apartments = unsdata.find_all('div',{'class':'sc_listingTile'})
    
    sleep(randint(2,5))
    
    for i in apartments:
        apart_title = i.find('div', {'class':'sc_listingTileArea'}).get_text()
        title.append(apart_title)

        apart_location = i.find('div', {'class':'sc_listingTileAddress'}).get_text()
        location.append(apart_location)

        prices = i.select('.sc_panel.sc_listingTile .sc_listingTileContent .sc_listingTilePrice span')
        apart_price = prices[0].get_text()
        price.append(apart_price)

        apart_rate = i.find('span', {'class':'sc_listingTilePriceRentTermDescription'}).get_text()
        rate.append(apart_rate)

        bedsNbaths = i.select('.sc_panel.sc_listingTile .sc_listingTileContent .sc_listingTileIcons span')
        apart_bedrooms = bedsNbaths[0].get_text()
        bedrooms.append(apart_bedrooms)

        apart_bathrooms = bedsNbaths[1].get_text()
        bathrooms.append(apart_bathrooms)
```

```python
apartments_df = pd.DataFrame({'title':title,
                             'location':location,
                             'bedrooms':bedrooms,
                             'bathrooms':bathrooms,
                             'price':price,
                             'rate':rate})
apartments_df['title'] = apartments_df['title'].str.strip().replace("\r\n","")

apartments_df['location'] = apartments_df['location'].str.strip().replace("\r\n","")

apartments_df['price'] = apartments_df['price'].str.strip().replace("\r\n","")
apartments_df['price'] = apartments_df['price'].str.replace('KSh','')

apartments_df['rate'] = apartments_df['rate'].str.strip().replace("\r\n","")
```

```python
apartments_df.to_csv('apartments.csv') #saving our dataframe to a csv file
```
